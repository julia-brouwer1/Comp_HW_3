```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def equation(x, y): 
    f = y**2 + 1
    return f
def true(x):
    f = math.tan(x)
    return f

x0 = 0
y0 = 0
h = 0.2
steps = 6
```

```python
def euler (x0, y0, h, steps):
    x = x0
    y = y0
    x_euler = []
    y_euler = []
    y_true = []
    print("xi\t yi\t\t slope\t\t true value\t error")
    for i in range(steps):
        x_euler.append(x)
        y_euler.append(y)
        y_true.append(true(x))
        
        t = true(x)
        e = true(x) - y
        slope = equation(x, y)
        print(f"{x:.1f}\t {y:.6f}\t {slope:.6f}\t {t:.6f}\t {e:.6f}")
        y = y + h * slope
        x = x + h
        
        
    #return (slope, xn, yn, t, e)
```

```python
euler(x0, y0, h, steps)
```

```python
def rk2 (x0, y0, h, steps):
    x = x0
    y = y0
    x_rk2 = []
    y_rk2 = []
    y_true_rk2 = []
    print("xi\t yi\t\t true\t\t error")
    for i in range(steps):
        x_rk2.append(x)
        y_rk2.append(y)
        y_true_rk2.append(true(x))

        t = true(x)
        e = true(x) - y
        print(f"{x:.1f}\t {y:.6f}\t {t:.6f}\t {e:.6f}")
        
        k1 = equation(x, y)
        k2 = equation(x + h, y + h * k1)
        y = y + 0.5 * h * (k1 + k2)
        x = x + h
```

```python
rk2 (x0, y0, h, steps)
```

```python
plt.figure()
plt.plot(x_euler, y_euler, marker='o', label= 'Euler Method')
plt.plot(x_rk2, y_rk2, marker='^', label='2nd Order Runge-Kutta Method')
plt.plot(x_euler, y_true, marker='v', label='True Value')
plt.title("Numerical Integrations")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
euler(x0, y0, .1, 20)
rk2 (x0, y0, .1, 20)
```

```python
#Comparing errors for steps of different sizes using the Euler Method
euler(x0, y0, .05, steps)
euler(x0, y0, .1, steps)
euler(x0, y0, .5, steps)
```

```python
#Comparing errors for steps of different sizes using the Runge-Kutta Method
rk2 (x0, y0, .1, steps)
rk2 (x0, y0, .5, steps)
rk2 (x0, y0, .7, steps)
```
