```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def equation(x, y): 
    f = y**2 + 1
    return f
def true(x):
    f = math.tan(x)
    return f

x0 = 0
y0 = 0
h = 0.2
steps = 6

```

```python
def euler (x0, y0, h, steps):
    x = x0
    y = y0
    x_euler = []
    y_euler = []
    y_true = []
    print("xi\t yi\t\t slope\t\t true value\t error") #uncomment if want to see table
    for i in range(steps):
        x_euler.append(x)
        y_euler.append(y)
        y_true.append(true(x))
        
        t = true(x)
        e = abs(true(x) - y)
        slope = equation(x, y)
        print(f"{x:.1f}\t {y:.6f}\t {slope:.6f}\t {t:.6f}\t {e:.6f}") #to see plot of values
        y = y + h * slope
        x = x + h
    return e, x_euler, y_euler, y_true
```

```python
def rk2 (x0, y0, h, steps):
    x = x0
    y = y0
    x_rk2 = []
    y_rk2 = []
    y_true_rk2 = []
    print("xi\t yi\t\t true\t\t error") 
    for i in range(steps):
        x_rk2.append(x)
        y_rk2.append(y)
        y_true_rk2.append(true(x))

        t = true(x)
        e = abs(true(x) - y)
        print(f"{x:.1f}\t {y:.6f}\t {t:.6f}\t {e:.6f}")
        
        k1 = equation(x, y)
        k2 = equation(x + h, y + h * k1)
        y = y + 0.5 * h * (k1 + k2)
        x = x + h
    return e, x_rk2, y_rk2, y_true_rk2
```

```python
#Convergence Study

def euler_CS (x0, y0, h, steps):
    x = x0
    y = y0
    for i in range(steps):
        slope = equation(x, y)
        y = y + h * slope
        x = x + h
    return y

def rk2_CS (x0, y0, h, steps):
    x = x0
    y = y0
    for i in range(steps):        
        k1 = equation(x, y)
        k2 = equation(x + h, y + h * k1)
        y = y + 0.5 * h * (k1 + k2)
        x = x + h
    return y
```

```python
x_final = 1.5
h_values = np.array([0.4, 0.2, 0.1, 0.05, 0.025])
x0 = 0
y0 = 0

h_ref = h_values[-1]
steps_ref = int((x_final - x0) / h_ref)

y_ref_euler = euler_CS(x0, y0, h_ref, steps_ref)
y_ref_rk2 = rk2_CS(x0, y0, h_ref, steps_ref)
```

```python
frac_diff_euler = []
frac_diff_rk2 = []

for h in h_values[:-1]:
    steps = int((x_final - x0) / h)
    y_e = euler_CS(x0, y0, h, steps)
    y_r = rk2_CS(x0, y0, h, steps)
    frac_diff_euler.append(abs(y_e - y_ref_euler) / abs(y_ref_euler))
    frac_diff_rk2.append(abs(y_r - y_ref_rk2)/ abs (y_ref_rk2))
```

```python
plt.figure()

plt.loglog(h_values[:-1], frac_diff_euler, '-o', label = 'Euler')
plt.loglog(h_values[:-1], frac_diff_rk2, '-^', label = 'RK2')
plt.xlabel("Step size (h)")
plt.ylabel("Fractional difference")
plt.title("Convergence Study")
plt.legend()
plt.savefig("Convergence_Study.png")
plt.show()
```

```python
#Maxwell-Boltzmann Stuff
m = 1.67e-27
k = 1.380649e-23
def MB_function (T, k, m, v):
    f = (m / (2 * np.pi * k * T))**(3/2) * 4 * math.pi * v**2 * np.exp(- m * v**2/(2 * k * T))
    return f
v_values = np.linspace(0, 30000, 2000)
y_values = MB_function (10000, k, m, v_values)
```

```python
plt.figure()
plt.plot(v_values, y_values, label="MaxwellBoltzmann")
plt.title("MaxwellBoltzmann")
plt.xlabel("v")
plt.ylabel("f(v)")
plt.legend()
plt.savefig("MaxwellBoltzmann.png")
plt.show()
```

```python
eV_to_J = 1.602e-19
delta_E = 10.2 * eV_to_J #in Joules
m = 1.67e-27
v_min = np.sqrt(2 * delta_E / m)
print ("Min speed in m/s:", v_min)
```

```python
v_max = 500000
def integrate_trap(v, f):
    total = 0
    for i in range(len(v) - 1):
        total += 0.5 * (f[i] + f[i+1]) * (v[i+1] - v[i])
    return total
```

```python
T = 10000
k = 1.380649e-23
m = 1.67e-27

v_values = np.linspace(v_min, v_max, 100000)
f_values = MB_function (T, k, m, v_values)

fraction = integrate_trap(v_values, f_values)

print("Fraction of atoms energetic enough:", fraction)
```
