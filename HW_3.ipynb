```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def equation(y): 
    f = y**2 + 1
    return f
def true(x):
    f = math.tan(x)
    return f

x0 = 0
y0 = 0
h = 0.2
steps = 5

x_values = []
y_euler = []
y_true = []
```

```python
def euler (x0, y0, h, steps):
    x = x0
    y = y0
    print("xi\t yi\t\t slope\t\t true value\t error")
    for i in range(steps + 1):
        x_values.append(x)
        y_euler.append(y)
        y_true.append(true(x))
        
        t = true(x)
        e = true(x) - y
        slope = equation(y)
        print(f"{x:.1f}\t {y:.6f}\t {slope:.6f}\t {t:.6f}\t {e:.6f}")
        y = y + h * slope
        x = x + h
        
        
    #return (slope, xn, yn, t, e)
```

```python
euler(x0, y0, h, steps)
```

```python
plt.figure()
plt.plot(x_values, y_euler, marker='o', label= 'Euler Method')
plt.plot(x_values, y_true, marker='o', label='True Value')
plt.title("Numerical Integrations")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
